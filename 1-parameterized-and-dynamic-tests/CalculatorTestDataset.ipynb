```python
import numpy as np
import pandas as pd
```

```python
LOWEST = -10000
HIGHEST = 10000
NUMBER = 1000
```

```python
d0 = {'a': [0, 0, 1, 0, -1, 1, -1], 'b': [0, 1, 0, -1, 0, 1, -1]}
df = pd.DataFrame(data=d0)

rand_a = np.random.randint(LOWEST, HIGHEST, size=NUMBER)
rand_b = np.random.randint(LOWEST, HIGHEST, size=NUMBER)
d1 = {'a': rand_a, 'b': rand_b}
df0 = pd.DataFrame(data=d1)

df = df.append(df0)
```

```python
df["sum"] = df["a"] + df["b"]
df["difference"] = df["a"] - df["b"]
df["product"] = df["a"] * df["b"]
df["quotient"] = df["a"] // df["b"]
df["xor"] = df["a"] ^ df["b"]
```

```python
df.to_csv('dataset.csv', sep=';', index=False)
```
